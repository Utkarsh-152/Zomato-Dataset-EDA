```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df= pd.merge(df,df_country, on= 'Country Code', how='left')
```

```python
final_df.dtypes
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
# Top 3 countries in Pie Chart
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

### Observation
Maximum records in Descending Order:
1. India
2. United States
3. United Kingdom

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When rating is between 4.5-4.9-----> Excellent
2. When rating is between 4.0-4.4-----> Very Good
3. When rating is between 3.5-3.9-----> good
4. When rating is between 2.4-3.4-----> Average
5. When rating is between 1.8-2.4-----> Poor

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y = "Rating Count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y = "Rating Count", hue='Rating color',palette=['white', 'red', 'orange', 'yellow', 'green','green'],data=ratings)
```

## Observation
1. Not rated count is very high
2. Maximum number of rating is between 2.5 -3.4

```python
#count plot
sns.countplot(x="Rating color", data=ratings, palette=['white', 'red', 'orange', 'yellow', 'green','green'])
```
